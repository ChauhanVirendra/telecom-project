```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_data(path):
    df = pd.read_csv(path)
    df['date'] = pd.to_datetime(df['date'], format='%d/%m/%y %H:%M', errors='coerce')
    df = df.dropna(subset=['date'])
    return(df)

df = load_data("phone_data.csv")
print("1. Load the dataset into a pandas DataFrame and display the first 5 rows.\n ")
df.head()
```

```python
def data_info(df):
    print("2. Check the basic information about the dataset (data types, non-null counts). \n")
    print(df.info(),'\n')
    print("null value in each column : \n")
    print(df.isnull().sum(),'\n')
    print("3. Display the summary statistics of numerical columns.\n")
    print(df.describe(),'\n')
    
    print("4. Count the number of unique values in each column. \n")
    print(df.nunique())
data_info(df)
```

```python
def total_call_in_data(df):
    print("5. Find the total number of calls made in the dataset.\n")
    total_calls = df[df['item'] == 'call'].shape[0]
    print(total_calls)
total_call_in_data(df)
```

```python
def combine_call_duration(df):
    print("6. Calculate the total duration of all calls combined.\n")
    total_call_duration = df[df['item'] == 'call']['duration'].sum()
    print(total_call_duration)
    
combine_call_duration(df)
```

```python
def total_sms_in_data(df):
    print("7. Count how many SMS messages were sent.\n")
    sms_count = np.sum(df['item'] == 'sms')
    print(sms_count,)

total_sms_in_data(df)
    
```

```python
def avg_call_duration(df):
    print("9. Calculate the average duration of calls.\n")
    avg_duration = df[df['item'] == 'call']['duration'].mean()
    print(avg_duration)

avg_call_duration(df)
```

```python
def month_with_high_comm(df):
    print("10. Identify the month with the highest number of communications.\n")
    df['hour'] = df['date'].dt.hour
    most_active_month = df['month'].value_counts().idxmax()
    
    print(most_active_month)

month_with_high_comm(df)
```

```python
def higheat_call_in_day(df):
    print("12. Find the day with the highest number of calls.")
    hightest_call_by_day = df[df['item'] == 'call']['date'].value_counts().idxmax()
    print(hightest_call_by_day)

higheat_call_in_day(df)
```

```python
def data_using_vs(df):
    print("13. Calculate the percentage of data usage vs calls vs SMS.\n")

    call_count = (df['item'] == 'call').sum()
    sms_count = (df['item'] == 'sms').sum()
    data_count = (df['item'] == 'data').sum()
    total = call_count + sms_count + data_count
    call_percent = (call_count/total)*100
    sms_percent = (sms_count/total)*100
    data_percent = (data_count/total)*100

    print("\ncall ",call_percent,'\nsms : ',sms_percent,'\ndata : ',data_percent)

data_using_vs(df)
```

```python
def avg_call_for_each_net(df):
    print("15. Calculate the average call duration for each network provider.\n")
    avg_callDU_network = df.groupby('network')['duration'].sum()
    print(avg_callDU_network)

avg_call_for_each_net(df)
```

```python
def highest_avg_call_dur(df):
    print("16. Find the network with the highest average call duration.")
    calls_df = df[df['item'] == 'call']
    avg_call_by_each_nt = calls_df.groupby('network')['duration'].mean()
    highest_avg_network = avg_call_by_each_nt.idxmax()
    print(highest_avg_network)

highest_avg_call_dur(df)
```

```python
def comm_occurred_on_data(df):
    print("17. Count how many communications occurred on each date.")

    df['only_date'] = df['date'].dt.date
    comm_count_per_day = df.groupby('only_date').size().reset_index(name='communication_count')
    print(comm_count_per_day)
    
comm_occurred_on_data(df)    
```

```python
def calculate_comm_for_each_day(df):
    print("18. Calculate the total duration of communications for each day.")
    total_duration_each_day = df.groupby(df['date'].dt.date)['duration'].sum()
    print(total_duration_each_day)

calculate_comm_for_each_day(df)
```

```python
def most_activ_hour_of_comm(df):
    print("19. Find the most active hour of communication (assuming all occurred at same time for this dataset).\n")
    df['hour'] = df['date'].dt.hour
    hour_counts = df['hour'].value_counts()
    most_active = hour_counts.idxmax()
    print(most_active)

most_activ_hour_of_comm(df)
```

```python
def call_sms_data_plivot_table(df):
    print("20. Create a pivot table showing the count of each item type (call, sms, data) per network.\n")
    table =  pd.pivot_table(df, index='network', columns='item', aggfunc='size', fill_value=0)
    print(table)

call_sms_data_plivot_table(df)
```

```python
def avrage_of_2day(df):
    print("21. Calculate the rolling 2-day average of call durations.\n")
    df['date_only'] = pd.to_datetime(df['date']).dt.date
    calls = df[df['item'] == 'call'].groupby('date_only')['duration'].sum()
    avg = calls.rolling(2).mean()
    print(avg)

avrage_of_2day(df)
```

```python
def anomalies_in_call(df):
    print("22. Identify any anomalies in call durations (values beyond 3 standard deviations from mean).\n")
    
    calls_df = df[df['item'] == 'call']
    
    mean_duration =  calls_df['duration'].mean()
    std_duration =  calls_df['duration'].std()
    
    upper_limit = mean_duration + 3 * std_duration
    lower_limit = mean_duration - 3 * std_duration

    anomalies = calls_df[(calls_df['duration'] > upper_limit) | (calls_df['duration'] < lower_limit)]
    
    print(anomalies[['date', 'duration', 'network']])
    
anomalies_in_call(df)
```

```python
def datafram(df):
    print("23. Create a new DataFrame that shows the count and total duration of each item type per month.\n")
    df['month'] = df['date'].dt.month
    summary = df.groupby(['month', 'item']).agg(
    count=('item', 'count'),
    total_duration=('duration', 'sum')
    ).reset_index()
    print(summary)

datafram(df)
```

```python
def cumulative(df):
    print("24. Calculate the cumulative sum of call durations over time.\n")
    
    df['date'] = pd.to_datetime(df['date'])
    calls = df[df['item'] == 'call'].sort_values('date')
    calls['cumulative_duration'] = calls['duration'].cumsum()
    return calls[['date', 'cumulative_duration']]

cumulative(df)
```

```python
def continuous_sequence(df):
    print("25. Find the longest continuous sequence of calls from the same network.\n")
    
    max_streak = (df['network'] != df['network'].shift()).cumsum()
    streak_counts = max_streak.value_counts().max()
    print(streak_counts)

continuous_sequence(df)
```

```python
def time_defference(df):
    print("26. Calculate the time difference between consecutive communications (requires datetime conversion).\n")
    
    df['date'] = pd.to_datetime(df['date'])
    df = df.sort_values('date')
    df['time_diff'] = df['date'].diff()
    calculation = df[['date', 'time_diff']]
    print(calculation)

time_defference(df)
```

```python
def high_low_activity(df):
    print('27. Create a function to categorize days as "high activity" or "low activity" based on number of communications.\n')
    
    comms = df.groupby(pd.to_datetime(df['date']).dt.date).size()
    threshold = comms.mean()
    formula = comms.apply(lambda x: 'high activity' if x > threshold else 'low activity')
    print(formula)
    
high_low_activity(df)
```

```python
def top_network(df):
    print("28. Calculate the network that was used most for each type of communication (call, sms, data).\n")
    
    highest_network = df.groupby('item')['network'].agg(lambda x: x.value_counts().idxmax())
    print(highest_network)

top_network(df)
```

```python
def call_and_message(df):
    print("29. Find the ratio of outgoing calls to SMS messages for each network.\n")
    
    calls = df[df['item'] == 'call'].groupby('network').size()
    sms = df[df['item'] == 'sms'].groupby('network').size()
    ratio = (calls / sms).fillna(0)
    print(ratio)

call_and_message(df)
```

```python
def summary_datafram(df):
    print("30. Create a summary DataFrame that shows for each network: total calls, total SMS, total duration, and average duration.\n")
    
    grouped = df.groupby('network').agg(
        total_calls=('item', lambda x: (x == 'call').sum()),
        total_sms=('item', lambda x: (x == 'sms').sum()),
        total_duration=('duration', 'sum'),
        avg_duration=('duration', 'mean')
    )
    print(grouped)

summary_datafram(df)
```
